## Just to make a fun visualization of Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('../data/advertising.csv')
df = pd.DataFrame(data)

df.head()
```

### With only one feature for now

```python
X = df[['TV']]
y = df[['Sales']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

lr.fit(X, y)
```

### Remember, our equation was something like this:
* y_hat = b + w*x_0

### How to find the bias (b)?

```python
lr.intercept_[0]
```

### How to find the weight of predictor variable?

```python
lr.coef_[0][0]
```

### Now we actually do have the equation, since x is our input:
* y_hat = 6.97 + 0.055*x

### Say, we want to predict the value where x = 500.

```python
round(lr.intercept_[0] + lr.coef_[0][0]*500, 2)
```

```python
plt.figure(figsize=(8, 6))

sns.regplot(x=X, y=y, scatter_kws={'color': 'black'}, color='red')
plt.title('Equation: 6.97 + 0.055*x')
plt.show()
```

### Let's recreate the function

```python
def f(x):
    return lr.intercept_[0] + lr.coef_[0][0]*x
```

```python
plt.figure(figsize=(8, 6))

l = []

for x in np.arange(0, 300, 1):
    l.append(f(x))

plt.scatter(X, y, c='black')
plt.plot(l, c='red')
plt.ylabel('Sales')
plt.xlabel('TV')
plt.title('Equation: 6.97 + 0.055*x')
plt.show()
```
